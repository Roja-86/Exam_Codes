##### Import Essential Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

##### Defining f(u) and u(k)

```python
# Define the nonlinear function of the plant
def f(u):
    return 0.6 * np.sin(np.pi * u) + 0.3 * np.sin(3 * np.pi * u) + 0.1 * np.sin(5 * np.pi * u)

# Define the input function u(k)
def input_u(k):
    return np.sin(2 * np.pi * k / 250)

# Range for k
k_values = np.arange(300, 701)

# Compute the function values
f_values = [f(input_u(k)) for k in k_values]
```

##### Plotting

```python
# Plot
plt.figure(figsize=(10, 5))
plt.plot(k_values, f_values, label='f(u(k))')
plt.xlabel('Time step k')
plt.ylabel('f(u(k))')
plt.title('Plot of f(u(k)) for 300 < k < 700')
plt.legend()
plt.grid(True)
plt.show()
```
